```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
house_df = pd.read_csv("trimmed_df_3.csv")
house_df.replace(to_replace = {'LivingArea': {0: 'nan'}})
pd.set_option('display.max_columns', 500)

price_loc = house_df.plot(kind = "scatter", x= "Latitude",y= "Longitude",
            c= "ClosePrice",cmap = 'plasma', s=19)

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("ClosePrice as Related to Location")
price_loc.set_facecolor("black")
price_loc
```

```python

year_loc = house_df.plot(kind = "scatter", x= "Latitude",y= "Longitude",
            c= "YearBuilt",cmap = 'plasma', s=19)

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Year Built as Related to Location")
year_loc.set_facecolor("black")
year_loc
```
